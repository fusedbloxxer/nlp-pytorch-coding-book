```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

corpus = [
    'Time flies flies like an arrow.',
    'Fruit flies like a banana.'
]
vocab = ['an', 'arrow', 'banana', 'flies', 'fruit', 'like', 'time']
one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(one_hot, annot=True, cbar=False, xticklabels=vocab, yticklabels=['Sentence 1', 'Sentence 2'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vectorizer = TfidfVectorizer()
tf_idf = tf_idf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tf_idf, cbar=False, annot=True, linewidths=.5, xticklabels=vocab, yticklabels=['Sentence 1', 'Sentence 2'])
```

```python
import torch
from torch import Tensor

def describe(x: Tensor) -> None:
    print('Tensor Type: {}\n'.format(x.type()))
    print('Tensor Shape: {}\n'.format(x.size()))
    print('Tensor Content: {}\n'.format(x))

describe(torch.zeros((2, 3)))
```
